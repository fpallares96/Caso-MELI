## ¿Cuál es la duración promedio de las ofertas? ¿Y la mediana de cuanto es?

```python
# Convertir las columnas de fecha y hora a formato datetime
df['OFFER_START_DTTM'] = pd.to_datetime(df['OFFER_START_DTTM'])
df['OFFER_FINISH_DTTM'] = pd.to_datetime(df['OFFER_FINISH_DTTM'])

# Calcular la duración de cada oferta en horas
df['OFFER_DURATION_HOURS'] = (df['OFFER_FINISH_DTTM'] - df['OFFER_START_DTTM']).dt.total_seconds() / 3600

# Calcular la duración promedio de las ofertas
media = df['OFFER_DURATION_HOURS'].mean()
mediana = df['OFFER_DURATION_HOURS'].median()
std = df['OFFER_DURATION_HOURS'].std()
```

Vemos que las ofertas duran en promedio 5.7 hs, la mediana de 6 hs y el desvío de 2.11 hs.

## ¿Hay algún horario en el cual las publicaciones fueron más exitosas?

```python
# Extraer la hora de inicio de cada oferta
df['OFFER_START_HOUR'] = df['OFFER_START_DTTM'].dt.hour

# Agrupar por hora de inicio y sumar las cantidades y montos vendidos
exito_por_hora = df.groupby('OFFER_START_HOUR').agg({'SOLD_QUANTITY': 'sum', 'SOLD_AMOUNT': 'sum'}).reset_index()

# Ordenar por éxito
exito_por_hora.sort_values(by=['SOLD_QUANTITY', 'SOLD_AMOUNT'], ascending=False, inplace=True)

exito_por_hora.head()
```

Evidendemente las ofertas lanzadas en estos horarios fueron las más exitosas. Pero esto se cumple para todas las verticales en cualquier momento?

```python
# Convertir las columnas de fecha y hora a formato datetime
df['OFFER_START_DTTM'] = pd.to_datetime(df['OFFER_START_DTTM'])

# Extraer mes, semana y hora
df['MES'] = df['OFFER_START_DTTM'].dt.month
df['SEMANA'] = df['OFFER_START_DTTM'].dt.isocalendar().week
df['HORA'] = df['OFFER_START_DTTM'].dt.hour

# Agrupar por mes, semana, vertical y hora
resultado = df.groupby(['MES', 'SEMANA', 'VERTICAL', 'HORA']).agg({'SOLD_QUANTITY': 'sum', 'SOLD_AMOUNT': 'sum'})

# Resetear el índice para facilitar el análisis
resultado.reset_index(inplace=True)

# Identificar las horas más exitosas por mes, semana y vertical
resultado.sort_values(by=['MES', 'SEMANA', 'VERTICAL', 'SOLD_QUANTITY', 'SOLD_AMOUNT'], ascending=[True, True, True, False, False], inplace=True)
```

```python
# Convertir las columnas de fecha y hora a formato datetime
df['OFFER_START_DTTM'] = pd.to_datetime(df['OFFER_START_DTTM'])

# Extraer la fecha (día) y la hora
df['FECHA'] = df['OFFER_START_DTTM'].dt.date
df['HORA'] = df['OFFER_START_DTTM'].dt.hour

# Calcular la hora más exitosa por día y por vertical
hora_exitosa = df.groupby(['FECHA', 'VERTICAL', 'HORA']).agg({'SOLD_QUANTITY': 'sum'}).reset_index()
hora_exitosa = hora_exitosa.sort_values(by=['FECHA', 'VERTICAL', 'SOLD_QUANTITY'], ascending=[True, True, False])
hora_exitosa = hora_exitosa.drop_duplicates(subset=['FECHA', 'VERTICAL'])

# Obtener una lista de las verticales únicas
verticales = hora_exitosa['VERTICAL'].unique()

# Crear un gráfico de líneas para cada vertical
for vertical in verticales:
    df_vertical = hora_exitosa[hora_exitosa['VERTICAL'] == vertical]
    plt.figure(figsize=(12, 8))
    sns.lineplot(data=df_vertical, x='FECHA', y='HORA')
    plt.title(f'Hora Más Exitosa a lo Largo del Tiempo para {vertical}')
    plt.xlabel('Fecha')
    plt.ylabel('Hora Más Exitosa')
    plt.xticks(rotation=45)
    plt.show()
```

## ¿Está sucediendo algo los fines de semana?

```python
# Convertir la columna de fecha a datetime
df['OFFER_START_DTTM'] = pd.to_datetime(df['OFFER_START_DTTM'])

# Función para clasificar si es fin de semana o día de semana
df['DIA_TIPO'] = df['OFFER_START_DTTM'].dt.dayofweek.apply(lambda x: 'Fin de Semana' if x >= 5 else 'Día de Semana')

# Agrupar por tipo de día y sumar las ventas
ventas_por_dia = df.groupby('DIA_TIPO')['SOLD_QUANTITY'].sum()

# Mostrar los resultados
print(ventas_por_dia)
```

```python
# Convertir la columna de fecha a datetime y extraer el día de la semana
df['OFFER_START_DTTM'] = pd.to_datetime(df['OFFER_START_DTTM'])
df['DIA_TIPO'] = df['OFFER_START_DTTM'].dt.dayofweek.apply(lambda x: 'Fin de Semana' if x >= 5 else 'Día de Semana')

# Agrupar por vertical y tipo de día, y sumar las ventas
ventas_por_dia_y_vertical = df.groupby(['VERTICAL', 'DIA_TIPO'])['SOLD_QUANTITY'].sum().reset_index()

# Obtener una lista de las verticales únicas
verticales = df['VERTICAL'].unique()

# Crear un gráfico de barras para cada vertical
for vertical in verticales:
    df_vertical = ventas_por_dia_y_vertical[ventas_por_dia_y_vertical['VERTICAL'] == vertical]
    plt.figure(figsize=(10, 6))
    sns.barplot(data=df_vertical, x='DIA_TIPO', y='SOLD_QUANTITY', palette="vlag")
    plt.title(f'Ventas por Tipo de Día para {vertical}')
    plt.xlabel('Tipo de Día')
    plt.ylabel('Cantidad Vendida')
    plt.show()
```

Evidentemente para todas las líneas de producto las mayores cantidad de ventas se producen los días de semana.

## En los días de semana, ¿Hay algun día que destaque sobre el resto?

```python
# Convertir la columna de fecha a datetime y extraer el día de la semana
df['OFFER_START_DTTM'] = pd.to_datetime(df['OFFER_START_DTTM'])
df['DIA_SEMANA'] = df['OFFER_START_DTTM'].dt.day_name()

# Definir el orden de los días de la semana
orden_dias = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['DIA_SEMANA'] = pd.Categorical(df['DIA_SEMANA'], categories=orden_dias, ordered=True)

# Agrupar por vertical y día de la semana, y sumar las ventas
ventas_por_dia_y_vertical = df.groupby(['VERTICAL', 'DIA_SEMANA'])['SOLD_QUANTITY'].sum().reset_index()

# Obtener una lista de las verticales únicas
verticales = df['VERTICAL'].unique()

# Crear un gráfico de barras para cada vertical
for vertical in verticales:
    df_vertical = ventas_por_dia_y_vertical[ventas_por_dia_y_vertical['VERTICAL'] == vertical]
    plt.figure(figsize=(12, 6))
    sns.barplot(data=df_vertical, x='DIA_SEMANA', y='SOLD_QUANTITY', palette="muted")
    plt.title(f'Ventas por Día de la Semana para {vertical}')
    plt.xlabel('Día de la Semana')
    plt.ylabel('Cantidad Vendida')
    plt.xticks(rotation=45)
    plt.show()
```

Vemos que en casi todas las verticales, se cumple que la cantidad de ventas comienzan en valores relativamente altos, continúan aumentando hasta el miercoles en donde encuentra sus máximos. Finalmente la tendencia comienza a descender hasta llegar a sus mínimos en Domingos.

## ¿Existe algún horario y día que me proporcione mayor probabilidad de éxito de acuerdo a mi vertical?
Entendiendo éxito como obtener mayores cantidades de ventas.

```python
# Convertir las columnas de fecha y hora a formato datetime
df['OFFER_START_DTTM'] = pd.to_datetime(df['OFFER_START_DTTM'])

# Extraer día de la semana y hora
df['DIA_SEMANA'] = df['OFFER_START_DTTM'].dt.day_name()
df['HORA'] = df['OFFER_START_DTTM'].dt.hour

# Agrupar por vertical, día de la semana y hora, y calcular ventas totales
ventas_por_vertical_dia_hora = df.groupby(['VERTICAL', 'DIA_SEMANA', 'HORA']).agg({'SOLD_QUANTITY': 'sum'}).reset_index()

# Identificar la mejor combinación para cada vertical
mejores_combinaciones = ventas_por_vertical_dia_hora.loc[ventas_por_vertical_dia_hora.groupby('VERTICAL')['SOLD_QUANTITY'].idxmax()]

print(mejores_combinaciones)
```
